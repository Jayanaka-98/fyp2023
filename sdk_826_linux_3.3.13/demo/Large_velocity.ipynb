```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_data = np.loadtxt('/home/linux/Documents/fyp2023-1/Data/Large_Motor_26.txt',delimiter=",")
```

```python
timestamp = input_data[:,0]
initial_timestamp = timestamp[0]
timestamp = timestamp - initial_timestamp
timestamp
```

```python
plt.figure(figsize=(20,10))

plt.plot(timestamp, input_data[:,1], c='r', label = 'X_ref',linewidth = 1)
plt.plot(timestamp, input_data[:,2], c='b', label = 'X_res',linewidth = 1)

plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs Time')

#plt.xlim(0.4e7,0.6e7)
#plt.ylim(0,2.5)
plt.legend()
plt.grid()

plt.show()
```
