```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_data = np.loadtxt('F_1.txt',delimiter=",")
```

```python
timestamp = input_data[:,0]
initial_timestamp = timestamp[0]
timestamp = timestamp - initial_timestamp
timestamp
```

```python
plt.figure(figsize=(20,10))

plt.plot(timestamp, -input_data[:,2], c='b', label = 'Reproduced Surface Variation',linewidth = 1)
plt.plot(timestamp, input_data[:,1], c='r', label = 'Tracked Surface Variation',linewidth = 1)


plt.xlabel('Time (s)')
plt.ylabel('Position (mm)')
plt.title('Position vs Time')

#plt.xlim(0,10.5)
#plt.xlim(0.6,0.8)
plt.legend()
plt.grid()

plt.show()
```
