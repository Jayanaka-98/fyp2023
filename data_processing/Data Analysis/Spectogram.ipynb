```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# Generate some random time series data
time = np.linspace(0, 10, 1000)
data = np.sin(5 * 2 * np.pi * time)
```

```python
# Define the parameters for the spectrogram
nperseg = 256  # Number of data points per segment
noverlap = 128  # Number of overlapping points between segments
```

```python
# Compute the spectrogram
frequencies, times, Sxx = signal.spectrogram(data, fs=1.0/(time[1]-time[0]), nperseg=nperseg, noverlap=noverlap)
```

```python
# Plot the spectrogram
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx))  # Convert to dB scale
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar(label='Power Spectral Density (dB)')
plt.title('Spectrogram')
plt.show()
```
